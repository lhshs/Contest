```python
import folium
import pandas as pd
```

```python
snow = pd.read_csv('./적설관측자료.csv', index_col = 0)
snow
```

```python
snow[snow['STN.1'].unique()]  #, 'LON', 'LAT']
```

```python
unique_loc = snow[['STN.1', 'LON', 'LAT']].drop_duplicates()
unique_loc
```

```python
korea_map = folium.Map(location=[38.250850, 128.564720], # 위도(LAT), 경도(LON)
                      zoom_start = 7)
korea_map
```

```python
for name, lat, lon in zip(unique_loc['STN.1'], unique_loc['LAT'], unique_loc['LON']):
    folium.Marker([lat, lon], popup = name, icon=folium.Icon(color='red')).add_to(korea_map)
korea_map
```

```python
korea_map.save('snow_loc.html')
```
