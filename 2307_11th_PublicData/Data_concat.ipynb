```python
path = 'D:/Contest/public_data/'

df1 = pd.read_csv(path + '2012.csv', encoding='cp949')
print(len(df1))
df2 = pd.read_csv(path + '2013.csv', encoding='cp949')
print(len(df2))
df3 = pd.read_csv(path + '2014.csv', encoding='cp949')
print(len(df3))
df4 = pd.read_csv(path + '2015.csv', encoding='cp949')
print(len(df4))
df5 = pd.read_csv(path + '2016.csv', encoding='cp949')
print(len(df5))
df6 = pd.read_csv(path + '2017.csv', encoding='cp949')
print(len(df6))
df7 = pd.read_csv(path + '2018.csv', encoding='cp949')
print(len(df7))
df8 = pd.read_csv(path + '2019.csv', encoding='cp949')
print(len(df8))
df9 = pd.read_csv(path + '2020.csv', encoding='cp949')
print(len(df9))
df10 = pd.read_csv(path + '2021.csv', encoding='cp949')
print(len(df10))
df11 = pd.read_csv(path + '2022.csv', encoding='cp949')
print(len(df11))
df12 = pd.read_csv(path + '2023.csv', encoding='cp949')
print(len(df12))

# for i in range(1, 13): print(len(f'df{i}'))
```
