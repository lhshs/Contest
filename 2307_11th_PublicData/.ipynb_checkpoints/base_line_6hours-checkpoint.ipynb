```python
df =pd.read_csv('한국전력거래소_5분단위 전력수급현황_20220407.csv',encoding='cp949')
```

```python
df_1=df[['기준일시','현재수요(MW)']]
```

```python
#시계열
def resid(df):
    date_resid=STL(df[['기준일시','현재수요(MW)']].set_index('기준일시'), period=12)
    df['resid']=date_resid.fit().resid.values
    return df
def trend(df):
    date_resid=STL(df[['기준일시','현재수요(MW)']].set_index('기준일시'), period=12)
    df['trend']=date_resid.fit().trend.values
    return df
def season(df):
    date_resid=STL(df[['기준일시','현재수요(MW)']].set_index('기준일시'), period=12)
    df['season']=date_resid.fit().seasonal.values
    return df
# 날짜
def date_columns(df):
    #날짜 컬럼 추가
    df['기준일시']=pd.to_datetime(df['기준일시'])
    df['year'] = df['기준일시'].dt.year
    df['month'] = df['기준일시'].dt.month
    df['day'] = df['기준일시'].dt.day
    df['weekday'] = df['기준일시'].dt.weekday
    return df
```

```python
df_1 = resid(df_1)
df_1 = trend(df_1)
df_1 = season(df_1)
df_1 = date_columns(df_1)
```

```python
df_1
```

```python

def forcast_next(df_1,next_num):
    df_1['next_target']=0
    for index in range(len(df_1)):
        try:
            df_1[f'next_target'][index] = df_1['현재수요(MW)'][index+next_num]
        except:continue
    df_1=df_1.iloc[:-next_num]
    return df_1
def split_train_and_test(df, date):
    """
    Dataframe에서 train_df, test_df로 나눠주는 함수
    df : 시계열 데이터 프레임
    date : 기준점 날짜
    """
    train = df[df['기준일시'] < date]
    test = df[df['기준일시'] >= date]
    del train['기준일시']
    del test['기준일시']
    y_train = train.pop(f'next_target')
    x_train=train.copy()
    y_test = test.pop(f'next_target')
    x_test=test.copy()
    return x_train,y_train,x_test,y_test
```

```python
result = []
for i in range(1,73):
    df_2 = df_1.copy()
    df_2 = forcast_next(df_2,i)
    x_train,y_train,x_test,y_test= split_train_and_test(df_2,'2021-01-07 23:35')
    lgb = LGBMRegressor()
    lgb.fit(x_train,y_train)
    predict = lgb.predict(x_test)
    result.append(mean_absolute_percentage_error(y_test,predict))
    # joblib.dump(lgb, f'lgb_{i}.pkl')
    print(i)
```

```python
result
```

```python
result_df = pd.DataFrame(result)
```
