### 기준일시 컬럼 변경

```python
import pandas as pd

# 데이터 포맷 변환 함수 정의
def convert_datetime_format(datetime_str):
    datetime_str = str(datetime_str)  # 정수형을 문자열로 변환
    year = datetime_str[:4]
    month = datetime_str[4:6]
    day = datetime_str[6:8]
    hour = datetime_str[8:10]
    minute = datetime_str[10:12]
    return f"{year}-{month}-{day} {hour}:{minute}"


# 기준일시 컬럼 포맷 변환
df_time['기준일시'] = df_time['기준일시'].apply(convert_datetime_format)
df_day['기준일시'] = df_day['기준일시'].apply(convert_datetime_format)
# 결과 출력
df_time['기준일시']
```

### 결측치 처리


- 결측치가 주로 -9 또는 -99로 되어 있다. 이를 None으로 변경하는 작업


```python
'''
df_time
제거 : 1 6 18 24 28
문자형 : 24 28
애매 : 4 10 18
-9  : 2 3 5 6 7 8 9 13 14 15 16 17 19 20 21 22 23 25  26 27 29 30 31 32 33 34 35 41 42 43 44 45
-99 : 11 12 37 38  39 40
'''
for i in range(len(df_time.columns)):
  print(i,":",df_time.columns[i])
  print(df_time[df_time.columns[i]].value_counts().sort_index())
  print('데이터 개수 :',df_time[df_time.columns[i]].value_counts().sort_index().sum())
  print('-'*50)
```

```python
# 결측치 -9 or -99 -> None
chg_9  = [ 2 ,3, 5, 6, 7, 8, 9,10, 13 ,14, 15, 16,17, 19, 20 ,21, 22 ,23 ,25,  26 ,27 ,29, 30, 31, 32 ,33, 34, 35, 41 ,42 ,43, 44 ,45]
chg_99 = [ 11, 12 ,37, 38 , 39 ,40 ]

for i in range(len(df_time.columns)):
  if i in chg_9:
    df_time[df_time.columns[i]] = df_time[df_time.columns[i]].apply(lambda x: None if x == -9 else x)
  if i in chg_99:
    df_time[df_time.columns[i]] = df_time[df_time.columns[i]].apply(lambda x: None if x == -99 else x)
  if i == 4 :
    df_time[df_time.columns[i]] = df_time[df_time.columns[i]].apply(lambda x: None if x == -6 else x)
    df_time[df_time.columns[i]] = df_time[df_time.columns[i]].apply(lambda x: None if x == -9 else x)

# 컬럼 제거
remove_cols = [1, 6, 18, 24, 28]  # 삭제할 컬럼의 인덱스 리스트

# 컬럼 인덱스로 컬럼 삭제
df_time.drop(df_time.columns[remove_cols], axis=1, inplace=True)
```

### groupby

```python
df_time_groupby = df_time.groupby('기준일시').mean()
df_day_groupby = df_day.groupby('기준일시').mean()

df_time_groupby = df_time_groupby.reset_index()
df_day_groupby = df_day_groupby.reset_index()
df_time_groupby
```

### 기준일시 누락 확인

```python
import pandas as pd

# 데이터프레임 생성 예시
df = df_time_groupby

# '기준일시' 컬럼을 datetime 형식으로 변환
df['기준일시'] = pd.to_datetime(df['기준일시'])

# 데이터프레임에서 최소 날짜와 최대 날짜 추출
min_date = df['기준일시'].min().date()
max_date = df['기준일시'].max().date()

# 전체 기간의 예상되는 데이터 개수
expected_count = (max_date - min_date).days * 24 + 1

# 실제 데이터 개수
actual_count = df['기준일시'].nunique()

# 누락된 데이터 확인 및 출력
expected_dates = pd.date_range(start=min_date, end=max_date, freq='H')
missing_dates = expected_dates[~expected_dates.isin(df['기준일시'])]
missing_data = pd.DataFrame({'누락된 데이터': missing_dates})
(missing_data)
```
