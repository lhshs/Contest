```python
# df1.drop(columns=['TM_FC', 'MOD', 'NE', 'STN', 'WF'], inplace=True)
df1 = df1.sort_values(by='TM_EF')
df1
```

```python
df1[df1['TM_EF'] == 201201010000]
```

```python
df1_group = df1.groupby('TM_EF').agg({'C': lambda x: x.mode()[0], 'W1': lambda x: x.mode()[0],
                                      'T': lambda x: x.mode()[0], 'W2': lambda x: x.mode()[0],
                                      'TA': lambda x: x.mode()[0], 'ST': lambda x: x.mode()[0],
                                      'SKY': lambda x: x.mode()[0], 'PREP': lambda x: x.mode()[0]})
df1_group
```

```python
df1_group.to_csv('단기육상예보_.csv')
```

```python
df2.drop(columns=['TM_FC', 'MOD', 'NE', 'STN', 'WF'], inplace=True)
df2
```

```python
df2_group = df2.groupby('TM_EF').agg({'C': lambda x: x.mode()[0], 'W1': lambda x: x.mode()[0],
                                      'T': lambda x: x.mode()[0], 'W2': lambda x: x.mode()[0],
                                      'S1': lambda x: x.mode()[0], 'S2': lambda x: x.mode()[0],
                                      'WH1': lambda x: x.mode()[0], 'WH2': lambda x: x.mode()[0],
                                      'SKY': lambda x: x.mode()[0], 'PREP': lambda x: x.mode()[0]})
df2_group
```

```python
df2_group.to_csv('단기해상예보_.csv')
```
